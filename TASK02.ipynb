**Cleaning Process**

```python
import numpy as pandas
from scipy import stats
```

```python
#detect the outliers using IQR
data2.columns
```

```python
Q1 = data2.quantile(0.25)
Q3 = data2.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data2=data2[~((data2<(Q1-1.5*IQR))|(data2>(Q3+1.5*IQR))).any(axis=1)]
```
